```python
import json
import pandas as pd
import time


fd = open("store_record.json",'r')
txt = fd.read()
fd.close()
record = json.loads(txt)

bill = []

print("WELCOME TO",'\033[1m' + 'STAR GENERAL STORE' + '\033[0m')
print("=============================\n")
print("Current offers : 10% Discount for total cost more than 1000 Rs\n                 15% Discount for total cost more than 3500 Rs\n                 30% Discount for total cost more than 7000 Rs")

print("......................................")
print("\nFill up purchaser' details--->")

name = str(input("\n Enter purchaser's name : "))
pno = str(input("Enter mobile number : "))

print("\n**********************************")
print('\033[1m' + 'THE ITEMS WE HAVE' + '\033[0m')
print("\n")
print('\033[1m' + "Purchase id   :    items" + '\033[0m')
print("---------------------------------")
for i in record:
    print(i, "         :   ",record[i]['item'])

print("***************************************\n")
while(True):
    pid_item = str(input("Enter unique purchase id : "))
    for i in record.keys():
        if(i == pid_item):
            k = 1
            break
        else:
            k = 0
    
    
    if(k!=0):
        print("Your order : " + record[pid_item]["item"])
        if(record[pid_item]['stock']==0):
            print("Sorry ! No stocks available")
        else:
            quant = int(input("Enter quantity of purchase : "))       
        
            if(quant > record[pid_item]['stock']):
                print("*--SORRY !! INSUFFICIENT STOCKS--*")
                print("Available stocks :",record[pid_item]['stock'])
                ch = str(input("Purchase all the stocks, Enter - Y/N --->"))
                if(ch == 'N' or ch == 'n'):
                    quant = 0
                    print("SORRY for the inconvenience")
                else:
                    quant = record[pid_item]['stock']
                    print("----Order placed successfully-----")
                    record[pid_item]['stock'] -= quant
                    p = record[pid_item]['price']
                    d = record[pid_item]['discount']
                    am = p*(1-(d/100))*quant
                    amount = round(am,2)

                    bill.append([record[pid_item]['item'], record[pid_item]['company'], quant, record[pid_item]['price'], record[pid_item]['discount'], amount])

            else:
                print("----Order placed successfully-----")
                record[pid_item]['stock'] -= quant
                p = record[pid_item]['price']
                d = record[pid_item]['discount']
                am = p*(1-(d/100))*quant
                amount = round(am,2)

                bill.append([record[pid_item]['item'], record[pid_item]['company'], quant, record[pid_item]['price'], record[pid_item]['discount'], amount])

                
        
    if(k==0):
        print("No such item id found !!")
        
    choice = str(input("\nPurchase more items ? Enter Y/N : "))
    if(choice == 'N' or choice == 'n'):
          break  
print("\n\n---------------------------------------------")
print('\033[1m' + 'ALL THE ORDERS HAVE BEEN PLACED SUCCESSFULLY' + '\033[0m')   
print("---------------------------------------------")
            
js_rec = json.dumps(record)

fd = open("store_record.json",'w')
fd.write(js_rec)
fd.close()



tot_am = 0

it = []
cm = []
pr = []
qu = []
dis = []
a = []


for i in bill:
    it.append(i[0])
    cm.append(i[1])
    pr.append(i[3])
    qu.append(i[2])
    dis.append(i[4])
    a.append(i[5])
    tot_am += i[5]
    
    
    
rec = {'Items' : it,'Company' : cm,'Price/item' : pr,'Quantity' : qu, 'Discount %' : dis,'Amount' : a}
tm = time.ctime()
df = pd.DataFrame(rec)

fp = open("sales_record.json",'r')
sl = fp.read()
fp.close()
sales = json.loads(sl)


print('\033[1m' + '\n\nPayment Slip' + '\033[0m')
print("BILL NO : ",len(sales) + 1)
print("-----------------------------------------------------------------------------------")  
print("Purchaser's name : ",name,"                   ",'time : ',tm)
print("Mobile no : " ,pno)
print('\n')
print(df)
print('\n')

if(tot_am > 1000 and tot_am < 3500):
    p_am = round(tot_am*(1-(10/100)),2)
elif(tot_am > 3500 and tot_am < 7000):
    p_am = round(tot_am*(1-(15/100)),2)
elif(tot_am > 7000):
    p_am = round(tot_am*(1-(30/100)),2)
else:
    p_am = tot_am
    
print("Total cost : ",tot_am, "Rs")
print('\033[1m' + 'Including offer, Total payble amount  :' + '\033[0m' ,p_am, "Rs")
print("-----------------------------------------------------------------------------------")  




bill_no = str(len(sales)+1)
sales[bill_no] = {'Name' : name, "Mobile No." : pno, "Purchase_record" : rec,'Total cost' : tot_am, 'Payble amount' : p_am,'Time' : tm}

js_sales = json.dumps(sales)
fd = open("sales_record.json",'w')
fd.write(js_sales)
fd.close()

```
