# Search Sales report by purchaser's Name:

```python
import json

fd = open("sales_record.json",'r')
txt = fd.read()
fd.close()
rec = json.loads(txt)

name = str(input("Enter name :"))

for i in rec:
    if(rec[i]['Name']==name):
        print("Bill no :",i)
        print(rec[i])
```
